```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5.0, 5.0, 0.1)

##slope and intercept can be adjusted to verify the changes in the graph
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

#plt.figure(figsize=(8,6))
plt.plot(x,ydata,'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#cubic
x = np.arange(-5.0, 5.0, 0.1)
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata= y + y_noise
plt.plot(x,ydata,'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#Quadratic
x = np.arange(-5.0, 5.0, 0.1)
y=np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise 
plt.plot(x, ydata, 'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#exponential
X = np.arange(-5.0, 5.0, 0.1)
Y = np.exp(X)
plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#logarithmic
X = np.arange(-5.0, 5.0, 0.1)
Y = np.log(X)
plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#sigmoidal/Logistic
X = np.arange(-5.0, 5.0, 0.1)
Y = 1-4/(1+np.power(3, X-2))
plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#non-linear regression example
import numpy as np
import pandas as pd

#download data
!wget -O china_gdp.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/china_gdp.csv
    
df = pd.read_csv('china_gdp.csv')
df.head(10)
```

```python
#plotting the data
plt.figure(figsize=(8,5))
x_data, y_data = (df['Year'].values, df['Value'].values)
plt.plot(x_data, y_data, 'ro')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
#choosing a model
X = np.arange(-5.0, 5.0, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))

plt.plot(X,Y)
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#buidling the mode
def sigmoid(x, Beta_1, Beta_2):
    y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
    return y
```

```python
beta_1 = 0.1
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(x_data,beta_1,beta_2)

#plot initial prediction against datapoints
plt.plot(x_data,Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
```

```python
#let's normalize our data
xdata = x_data/max(x_data)
ydata = y_data/max(y_data)
```

```python
#How to find the best parameters for our fit line?
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata,ydata)
#print the final parameters
print('beta_1 = %f, beta_2 = %f' % (popt[0],popt[1]))
```

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
#accuracy of the model
#split data into train/test
msk = np.random.rand(len(df)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

#build the model using train set
popt, pcov = curve_fit(sigmoid, train_x, train_y)

#predict using test set
y_hat = sigmoid(test_x, *popt)

#evaluation
print('Mean absolute error: %.2f' % np.mean(np.absolute(y_hat - test_y)))
print('Residual sum of suqares (MSE): %.2f' % np.mean((y_hat - test_y) ** 2))
from sklearn.metrics import r2_score
print('R2-score: %.2f' % r2_score(y_hat, test_y))
```
